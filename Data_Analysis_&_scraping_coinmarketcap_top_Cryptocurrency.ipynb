```python
import requests
from bs4 import BeautifulSoup

def get_top_cryptos():
    url = 'https://coinmarketcap.com/'
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        crypto_table = soup.find('table', {'id': 'cryptocurrencies'})

        if crypto_table:
            cryptos = []
            rows = crypto_table.find('tbody').find_all('tr')

            for row in rows:
                name = row.find('p', class_='coin-item-symbol').text.strip()
                rank = row.find('td', class_='text-center').text.strip()
                price = row.find('td', {'data-sort': 'price'}).text.strip()
                market_cap = row.find('td', {'data-sort': 'market_cap'}).text.strip()

                crypto_data = {
                    'Rank': rank,
                    'Name': name,
                    'Price': price,
                    'Market Cap': market_cap
                }
                cryptos.append(crypto_data)

            return cryptos
        else:
            print('Crypto table not found on the page.')
    else:
        print('Failed to retrieve the webpage.')

if __name__ == "__main__":
    top_cryptos = get_top_cryptos()

    if top_cryptos:
        for crypto in top_cryptos:
            print(crypto)
```

```python
import requests
from bs4 import BeautifulSoup

def list_tables(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        tables = soup.find_all('table')

        if tables:
            for index, table in enumerate(tables, 1):
                print(f"Table {index}:")
                print(table)
                print("\n" + "="*50 + "\n")
        else:
            print('Aucune table trouvée sur la page.')
    else:
        print('Échec de la récupération de la page web.')

# Remplacez l'URL par celle du site CoinMarketCap
url_coinmarketcap = 'https://coinmarketcap.com/'
list_tables(url_coinmarketcap)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_trending_cryptos():
    url = 'https://coinmarketcap.com/trending-cryptocurrencies/'
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        crypto_cards = soup.find_all('div', class_='iDhYBm')

        if crypto_cards:
            cryptos = []

            for card in crypto_cards:
                name = card.find('p', class_='coin-item-symbol').text.strip()
                rank = card.find('p', class_='hNpRff').text.strip()
                price = card.find('p', class_='price___3rj7O').text.strip()
                market_cap = card.find('p', class_='sc-1eb5slv-0 gGIpIK').text.strip()

                crypto_data = {
                    'Rank': rank,
                    'Name': name,
                    'Price': price,
                    'Market Cap': market_cap
                }
                cryptos.append(crypto_data)

            return cryptos
        else:
            print('No trending cryptocurrencies found on the page.')
    else:
        print('Failed to retrieve the webpage.')

# Utiliser la fonction et imprimer les résultats
trending_cryptos = scrape_trending_cryptos()

if trending_cryptos:
    for crypto in trending_cryptos:
        print(crypto)
```

```python
import requests

def get_trending_cryptos(api_key):
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
        'start':'1',
        'limit':'10',  # Vous pouvez ajuster le nombre de cryptos à récupérer
        'convert':'USD'
    }
    headers = {
        'Accepts': 'application/json',
        'X-CMC_PRO_API_KEY': api_key,
    }

    response = requests.get(url, headers=headers, params=parameters)

    if response.status_code == 200:
        data = response.json()
        cryptos = data['data']

        result = []

        for crypto in cryptos:
            rank = crypto['cmc_rank']
            name = crypto['name']
            price = crypto['quote']['USD']['price']
            market_cap = crypto['quote']['USD']['market_cap']

            crypto_data = {
                'Rank': rank,
                'Name': name,
                'Price': price,
                'Market Cap': market_cap
            }
            result.append(crypto_data)

        return result
    else:
        print('Failed to retrieve data from CoinMarketCap API.')
        print(response.text)
        return None

# Remplacez 'YOUR_API_KEY' par votre clé d'API réelle
api_key = 'f973c35c-bd10-46c2-8c27-4f619cb5d277'
trending_cryptos = get_trending_cryptos(api_key)

if trending_cryptos:
    for crypto in trending_cryptos:
        print(crypto)
```
